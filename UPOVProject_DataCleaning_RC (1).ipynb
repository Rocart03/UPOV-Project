```python
# Dropping all year columns with an N at the end,
# Defining function letter dropper
def letter_dropper(df):
    for col in df.columns: 
        if col.endswith ('N'):
            df.drop(col, axis=1, inplace= True)
```

```python
# Calling letter dropper with the crops df
letter_dropper(crops_df)
# Checkpoint
crops_df.head()
```

```python
# Combining flag with its associated value 
# Defining flag combiner 
def flag_combiner(df):
    for col in df.columns:
        if col.endswith('F'):
            year = col[1:5]
# Combining the values from the year column and flag column as strings
            df[year] = df[f'Y{year}'].astype(str) + df[col].astype(str)
# Dropping the original year and flag columns
            df.drop([f'Y{year}', col], axis=1, inplace=True)
```

```python
# Calling flag combiner 
flag_combiner(crops_df)
crops_df.head()
```

```python
# Spliting the flag code off of value and creating its own column 
def flag_converter(df, production_column):
# Creating a new column for production flags
    df['Production Flag'] = ''
# Iterateing through each row 
    for index, row in df.iterrows():
# Extracting the flag part of the production value
        flag = row[production_column][-1]
# Updating the production flag column with the extracted flag
        df.at[index, 'Production Flag'] = flag
# Removing the flag from the production values
    df[production_column] = df[production_column].str[:-1]
    return df
```

```python
# Calling flag converter
flag_converter(prod_df_long, 'Production (Tonnes)').head()
```

```python
# We previosuly anazlyzed which countries were missing data and excluded them from the analysis 
# Each of the countries we selected in East Africa (3 UPOV, 12 Upov) should have 61 rows for data from 1961-2021
# Some countries dont grow one of the crops so they would have no data on a specfic crop at all
# For each combination of area and item there should be 61 rows
# Defining row checker for each combo of area and item for validation of 61 rows
def row_checker(df):
# Grouping area and item and counting the number of rows in each group
    grouped = df.groupby(['Area', 'Item']).size().reset_index(name='Row Count')
# Checking if all groups have 61 rows or more
    all_61 = grouped['Row Count'].min() == 61
# If all groups have 61 rows print the correct number of rows
    if all_61:
        print("Correct number of rows")
        print(grouped)
    else:
        print("Not all groups have 61 rows or more.")
```

```python
row_checker(east_df)
```

```python
# Preparing data from aggregation columns for Annual Area harvested change and Annual production change.
# Sorting datafor each combination of area and item in chronological order by year
eastdf_sorted = east_df.sort_values(['Area', 'Item', 'Year']).reset_index(drop=True)
```

```python
eastdf_sorted.head()
```

```python
eastdf_sorted.info()
```

```python
# Converting area harvest and production columns to float
eastdf_sorted['Area Harvested (100 g/ha)'] = eastdf_sorted['Area Harvested (100 g/ha)'].astype(float)
eastdf_sorted['Production (Tonnes)'] = eastdf_sorted['Production (Tonnes)'].astype(float)
```

```python
# Defining annual area change 
# Will caluclate the annual difference between each item for each country 
def annual_area_change(df):
# Calculating the difference in area harvested between consecutive years
    df['Annual Area Harvested Change'] = df.groupby(['Area', 'Item'])['Area Harvested (100 g/ha)'].diff()
    return df
```

```python
# Calling function to calculate the annual area harvested change
new_east_df = annual_area_change(eastdf_sorted)
```

```python
new_east_df.head()
```

```python
# Repeating same steps for annual production change.
def annual_production_change(df):
# Calculating the difference in area harvested between consecutive years
    df['Annual Production Change'] = df.groupby(['Area', 'Item'])['Production (Tonnes)'].diff()
    return df
```

```python
# Calling function to calculate the annual production change
new_east_df2 = annual_production_change(new_east_df)
```

```python
# Both Aggregated columns for year 1961 will have NaNs because
# the data doesnt have a previous year to calculate the difference from since the data starts at 1961
# Converting those NaNs to zeros for those instances
new_east_df2['Annual Area Harvested Change'].fillna(0, inplace=True)
new_east_df2['Annual Production Change'].fillna(0, inplace=True)
```
