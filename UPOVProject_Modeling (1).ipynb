```python
# Import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)
```

```python
# reading in csv
file_path = "C:\\Users\\Graduate School\\Downloads\\east_africa_final.csv"
east_df = pd.read_csv(file_path)
east_df.head(10)
```

```python
east_df.info()
```

```python
east_df = east_df.rename(columns={
    'Yield Value (100 g/ha)': 'Area Harvested (100 g/ha)',
    'Yield Value Flag': 'Area Harvested Flag',
    'Production Value Flag': 'Production Flag',
    'Yield Difference': 'Annual Area Harvested Change',
    'Production Difference': 'Annual Production Change'
})

east_df.tail(1)
```

```python
import statsmodels.api as sm

# Creating a random effects model using the correct column names
model = sm.MixedLM.from_formula("Q('Annual Area Harvested Change') ~ Q('Entry Year')", east_df, groups=east_df["Area Code"])
result = model.fit()

# Printing the summary of the model
print(result.summary())
```
